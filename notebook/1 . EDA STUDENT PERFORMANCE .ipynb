### 3.5 Adding columns for "Total Score" and "Average"

```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head(10)
```

```python
# math_full = df[df['math_score'] == 100]['average'].count()


math_full = (df['math_score'] == 100).sum()
reading_full = (df['reading_score'] == 100).sum()
writing_full = (df['writing_score'] == 100).sum()

print('Number of students with full marks in Maths: ', math_full)
print('Number of students with full marks in Reading: ', reading_full)
print('Number of students with full marks in Writing: ', writing_full)
```

```python
# reading_less_20 = df[df['reading_score'] <= 20]['average'].count()


math_below_20 = (df['math_score'] <= 20).sum()
reading_below_20 = (df['reading_score'] <= 20).sum()
writing_below_20 = (df['writing_score'] <= 20).sum()

print('Number of students with ≤20 marks in Maths: ', math_below_20)
print('Number of students with ≤20 marks in Reading: ', reading_below_20)
print('Number of students with ≤20 marks in Writing: ', writing_below_20)
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

X = ['Total Average', 'Math Average']

# Access using index labels
female_scores = [gender_group.loc['female', 'average'], gender_group.loc['female', 'math_score']]
male_scores = [gender_group.loc['male', 'average'], gender_group.loc['male', 'math_score']]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female')

plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average vs Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (15, 9)
sns.set_theme(style='whitegrid')   

sns.countplot(data=df, x='lunch', palette='PuBu')
plt.title('Comparison of different types of lunch', fontweight=30, fontsize=20)
plt.xlabel('Types of Lunch')
plt.ylabel('Count')
plt.show()
```

#### Insights
